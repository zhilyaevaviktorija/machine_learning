## ЗАДАНИЕ 1: Базовые графики matplotlib

```python
# Создайте фигуру с 4 subplots (2x2):
# 1. Линейный график функции y = x^2 * sin(x) на интервале [0, 10]
#    Добавьте заголовок, подписи осей, сетку, легенду
# 2. Гистограмма 1000 случайных точек из нормального распределения
#    Настройте количество bins=30, добавьте заголовок
# 3. Scatter plot: 50 точек (x - равномерное распределение, y = 2x + шум)
#    Настройте размер точек, прозрачность, цвет
# 4. Bar chart: сравнение 5 разных моделей по точности
#    accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
#    Добавьте значения над столбцами

import numpy as np
import matplotlib.pyplot as plt

# Создаем фигуру с 4 subplots (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # 2 строки, 2 столбца

# Название всей фигуры
fig.suptitle('Базовые графики matplotlib', fontsize=16, y=1.02)


# График 1. Линейный график y = x^2 * sin(x)
ax = axes[0, 0]
x = np.linspace(0, 10, 100)
y = x**2 * np.sin(x)

ax.plot(x, y, 'b-', label='y = x^2 * sin(x)', linewidth=2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Линейный график')
ax.legend()
ax.grid(True, alpha=0.3)


# График 2. Гистограмма нормального распределения
ax = axes[0, 1]
data = np.random.normal(0, 1, 1000)

ax.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
ax.set_xlabel('Значение')
ax.set_ylabel('Частота')
ax.set_title('Гистограмма нормального распределения')


# График 3. Scatter plot: y = 2x + шум
ax = axes[1, 0]
x = np.random.uniform(0, 10, 50)
y = 2 * x + np.random.normal(0, 1, 50)
ax.scatter(x, y, alpha=0.6, color='green', edgecolors='darkgreen', linewidth=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Scatter plot: y = 2x + шум')
ax.grid(True, alpha=0.3)


# График 4. Bar chart: точность моделей
ax = axes[1, 1]
models = ['A', 'B', 'C', 'D', 'E']
accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
bars = ax.bar(models, accuracy, color=['skyblue', 'lightblue', 'deepskyblue', 'dodgerblue', 'steelblue'])
ax.set_xlabel('Модель')
ax.set_ylabel('Accuracy')
ax.set_title('Сравнение моделей')

# Добавляем значения над столбцами
for bar, acc in zip(bars, accuracy):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.005,
            f'{acc:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

## ЗАДАНИЕ 2: Seaborn визуализации

```python
# Используйте датасет tips из seaborn:
# import seaborn as sns
# tips = sns.load_dataset('tips')

# Создайте 3 графика:
# 1. Box plot: total_bill по дням недели (day)
#    Разделите дополнительно по времени (time) с помощью hue
# 2. Violin plot: tip в зависимости от размера компании (size)
# 3. Pair plot: выберите 4 числовых признака, раскрасьте по полу (sex)

import seaborn as sns
import matplotlib.pyplot as plt

# Загружаем датасет
tips = sns.load_dataset('tips')

# Создаем фигуру для графиков
plt.figure(figsize=(10, 11))

# График 1. Box plot: total_bill по дням с разделением по времени
# data=tips, x='day', y='total_bill', hue='time'
plt.subplot(2, 2, 1)
sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
plt.title('Total_bill по дням и времени')
plt.xlabel('День недели')
plt.ylabel('Total bill')
plt.grid(True, alpha=0.3)


# График 2. Violin plot: tip по размеру компании
# x='size', y='tip'
plt.subplot(2, 2, 2)
sns.violinplot(data=tips, x='size', y='tip', color='blue',
               fill=True, inner='quartile', linewidth=1.5)
# inner='quartile' - показывает квартили внутри
plt.title('Tip по размеру компании')
plt.xlabel('Размер компании')
plt.ylabel('Tip')
plt.grid(True, alpha=0.3)


# График 3. Pair plot: 4 числовых признака, раскраска по полу
# tips[['total_bill', 'tip', 'size', 'sex']]
# Создаем отдельную фигуру для pair plot
plt.figure(figsize=(10, 8))
sns.pairplot(tips[['total_bill', 'tip', 'size', 'sex']],
             hue='sex', palette={'Male': 'blue', 'Female': 'red'},
             plot_kws={'alpha': 0.6},
             diag_kind='hist')
plt.suptitle('Pair plot по полу', y=1.02)
plt.show()
```
